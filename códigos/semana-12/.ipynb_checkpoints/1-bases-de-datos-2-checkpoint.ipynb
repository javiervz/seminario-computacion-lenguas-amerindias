### ¿Cómo construimos un mapa con las lenguas de Sudamérica que aparecen en Wals?

```python
## lenguas de wals https://github.com/cldf-datasets/wals/tree/master/cldf
languages_wals = pd.read_csv('languages_wals.csv')
```

```python
languages_wals
```

```python
## filtramos por lenguas de Sudamérica
languages_wals = languages_wals[languages_wals['Macroarea'] == 'South America']
```

```python
languages_wals
```

```python
## extraemos tres listas: lenguas, latitud, longitud

## lenguas
lenguas = list(languages_wals['Name'])

## latitud
latitud = list(languages_wals['Latitude'])

## longitud
longitud = list(languages_wals['Longitude'])
```

```python
lenguas[:5], latitud[:5], longitud[:5]
```

```python
## construyamos un diccionario lengua:[latitud, longitud]
diccionario_loc = {}

## recorremos los índices
for i in range(len(lenguas)):
    ## lengua
    lengua = lenguas[i]
    
    ## latitud
    lat = latitud[i]
    
    ## longitud
    lon = longitud[i]
    
    ## guardamos
    diccionario_loc[lengua] = [lat, lon]
```

```python
diccionario_loc
```

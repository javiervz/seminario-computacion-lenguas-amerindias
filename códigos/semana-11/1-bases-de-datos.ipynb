### Pregunta 1: ¿Cuántas lenguas hay por macroarea?

```python
## identificamos las macroareas

## definimos una lista vacía
macroareas = []

## recorremos la lista de macroareas
for M in list(languages_glotto['macroarea']):
    ## si no está, la agregamos
    if M not in macroareas and str(M)!= 'nan':
        macroareas += [M]
```

```python
## verificamos

macroareas
```

```python
languages_glotto.columns
```

```python
## diccionario de macroareas
## keys - macroareas
## values - listas de lenguas asociadas a esa macroarea

macroareas_dict = {}

## recorremos las macroareas
for M in macroareas:
    ## filtramos por lenguas de una macroarea
    languages_macroarea = languages_glotto[languages_glotto['macroarea'] == M]
    ## agregamos
    macroareas_dict[M] = languages_macroarea
```

```python
## número de lenguas: tamaño de los dataframes

for M in macroareas_dict.keys():
    ## mostramos el largo de los dataframes
    print(M, len(macroareas_dict[M]))
```

### Pregunta 2: ¿Cuántas familias lingüísticas hay por macroarea?

```python
## https://glottolog.org/meta/downloads
languoid = pd.read_csv('languoid.csv')
```

```python
languoid.columns
```

```python
languoid
```

```python
languoid = languoid[['id', 'family_id']]
```

```python
languoid = languoid[languoid['id'].isin(lista_lenguas_glottolog)]
```

```python
languoid
```

```python
# ahora, filtramos las filas que tienen datos faltantes

#languoid = languoid.dropna()
```

```python
languoid
```

```python
languoid['family_id'] = languoid['family_id'].mask(languoid['family_id'].isna(), languoid['id'])
```

```python
languoid
```

```python
## diccionario lengua:familia

## lista lenguas
L = list(languoid['id'])

## lista familias
F = list(languoid['family_id'])

## diccionario vacío
language_family = {}

## recorremos las lenguas: usamos los índices
for i in range(len(L)):
    
    ## lengua
    lengua = L[i]
    
    ## familia
    familia = F[i]
    
    ## creamos el par lengua:familia
    language_family[lengua] = familia
```

```python
language_family['mose1249']
```

```python
## el diccionario macroareas_dict tiene como keys las macroareas y como values los dataframes
## creamos una versión con las lenguas

macroareas_dict_lenguas = {}

## recorremos las macroareas
for macroarea in macroareas_dict.keys():
    
    ## dataframe
    lenguas = macroareas_dict[macroarea]
    
    ## lenguas
    lenguas = list(lenguas['glottocode'])
    
    ## guardamos
    macroareas_dict_lenguas[macroarea] = lenguas
```

```python
macroareas_dict_lenguas['South America']
```

```python
## comprobamos! número de lenguas: tamaño de las listas

for M in macroareas_dict_lenguas.keys():
    ## mostramos el largo de las listas
    print(M, len(macroareas_dict_lenguas[M]))
```

### Pregunta 3: ¿Cuántas lenguas están anotadas en Wals por macroarea?

```python
## lenguas de wals https://github.com/cldf-datasets/wals/tree/master/cldf
languages_wals = pd.read_csv('languages_wals.csv')
```

```python
languages_wals
```

```python
## lista de lenguas de wals
lista_lenguas_wals = list(languages_wals['Glottocode'])

## creamos un diccionario con listas
macroareas_dict_lenguas_wals = {M:[] for M in macroareas_dict_lenguas.keys()}

## recorremos las macroareas de macroareas_dict_lenguas
for M in macroareas_dict_lenguas.keys():
    
    ## lenguas de una macroarea
    lenguas_macroarea = macroareas_dict_lenguas[M]
    
    ## revisamos cuáles de estas lenguas están en wals
    for lengua in lenguas_macroarea:
        ## si está en la lista de lenguas de wals
        if lengua in lista_lenguas_wals:
            ## guardamos
            macroareas_dict_lenguas_wals[M] += [lengua]
```

```python
## miramos! 

for M in macroareas_dict_lenguas.keys():
    ## mostramos el largo de las listas
    print(M, len(macroareas_dict_lenguas[M]), len(macroareas_dict_lenguas_wals[M]))
```

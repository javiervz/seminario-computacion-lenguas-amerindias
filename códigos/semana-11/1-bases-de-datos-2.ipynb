```python
L = list(wals['Parameter_ID'])

## número de filas
print(len(L))

## version 1: número de rasgos de wals
print(len(set(L)))

print(list(set(L)))

## versión 2
rasgos_wals = []
for rasgo in L:
    if rasgo not in rasgos_wals:
        rasgos_wals = rasgos_wals + [rasgo]
        
print(rasgos_wals)
```

```python
## rasgo:[lenguas anotadas]
D = {}

## recorremos los rasgos
for rasgo in list(set(L)):
    ## wals filtrado x rasgo
    wals_filtrado = wals[wals['Parameter_ID']==rasgo]
    ## lista de lenguas
    lenguas = list(wals_filtrado['Language_ID'])
    ## guardamos
    D[rasgo] = lenguas
```

```python
D_numero_lenguas = {}

for rasgo in D.keys():
    ## número de lenguas
    D_numero_lenguas[rasgo] = len(D[rasgo])
```

**Pregunta 2.** ¿Cómo organizamos la lista de lenguas de Sudamérica en familias lingüísticas?

```python
## archivo descargado de glottolog
languoid = pd.read_csv('languoid.csv')
```

```python
languoid
```

```python
## filtremos la columna "id" con las lenguas de la lista glottocode_lenguas_SA

languoid = languoid[languoid['id'].isin(glottocode_lenguas_SA)]
```

```python
languoid
```

```python
## arreglemos los NaN

## reemplazamos con un string 
languoid.fillna('unk', inplace=True)
```

```python
languoid
```

```python
## construyamos un diccionario glottocode:glottocode_familia

dict_glottocode_fam = {}

## usamos las dos primeras columnas

## recorremos las lenguas de la primera columna
for i in range(len(languoid['id'])):
    ## accedemos a la lengua
    lengua = list(languoid['id'])[i]
    ## accedemos a la familia
    familia = list(languoid['family_id'])[i]
    ## guardamos la información en el diccionario
    dict_glottocode_fam[lengua] = familia
```

```python
dict_glottocode_fam
```

```python
## identifiquemos las familias únicas

## lista vacía
familias_unicas = []

## recorremos las familias
for familia in dict_glottocode_fam.values():
    ## si no está, la guardamos
    if familia not in familias_unicas:
        ## guardamos
        familias_unicas += [familia]
```

```python
## juntemos las lenguas con la misma familia

## diccionario vacío
dict_fam_lenguas = {}

## definimos pares familia:[]
## recorremos las familias
for familia in familias_unicas:
    ## definimos los pares
    dict_fam_lenguas[familia] = []
    
## recorremos las lenguas
for lengua in dict_glottocode_fam.keys():
    ## guardamos las lenguas en las listas respectivas
    ## identificamos la familia
    familia = dict_glottocode_fam[lengua]
    ## guardamos
    dict_fam_lenguas[familia] += [lengua]
```

```python
dict_fam_lenguas
```

```python
## ¿Cómo podríamos usar diccionarios para representar esta información?

## un detalle sobre pandas dataframes
## con list() transformamos una columna en lista
lenguas = list(df_argex5_1['name'])
```

```python
lenguas[:10]
```

```python
## valores para los distintos rasgos

## argex5-1
valores_argex5_1 = list(df_argex5_1['value'])

## argex5-2
valores_argex5_2 = list(df_argex5_2['value'])

## argex5-3
valores_argex5_3 = list(df_argex5_3['value'])
```

```python
## ¿Qué diccionarios podrían construirse? 

## idea 1: tres diccionarios - argex5_X_dict = {lengua1:valor, lengua2:valor, ...}

## diccionario vacío rasgo argex5-1
argex5_1_dict = {}

## recorremos las lenguas
for i in range(len(lenguas)):
    ## extraemos el valor
    valor = valores_argex5_1[i]
    ## guardamos en el diccionario
    argex5_1_dict[lenguas[i]] = valor
```

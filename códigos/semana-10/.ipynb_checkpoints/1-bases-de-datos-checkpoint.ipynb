### Glottolog!
**Pregunta 1.** ¿Cómo obtenemos una lista de lenguas de Sudamérica?

```python
## ¿Cómo se ven los datos de Glottolog?

## importamos la librería
import pandas as pd

## archivo descargado de glottolog
glottolog = pd.read_csv('languages_and_dialects_geo.csv')
```

```python
## columnas del dataframe glottolog

glottolog.columns
```

```python
glottolog
```

```python
## ¿Cuáles son los valores de la columna "macroarea"?

## lista vacía para guardar las macroáreas
macroareas_unicas = []

## recorremos los datos de la columna
for m in glottolog['macroarea']:
    ## si no está, lo guardamos
    if m not in macroareas_unicas:
        ## guardamos en la lista
        macroareas_unicas += [m]
```

```python
macroareas_unicas
```

```python
## otra forma

macroareas_unicas = set(glottolog['macroarea'])
```

```python
macroareas_unicas
```

```python
## ¿Cómo filtramos las lenguas de Sudamérica?

## accedemos a la columna "macroarea", y filtramos las filas que tengan "South America"
glottolog = glottolog[glottolog['macroarea'] == 'South America']
```

```python
glottolog
```

```python
## ¿Hay 1164 lenguas en Sudamérica? Filtremos por ciertas columnas: glottocode, name, isocode, latitude, longitude

glottolog = glottolog[['glottocode', 'name', 'isocodes', 'latitude', 'longitude']]
```

```python
glottolog
```

```python
## ahora, filtramos las filas que tienen datos faltantes

glottolog = glottolog.dropna()
```

```python
## ¿Se ve razonable este número?

glottolog
```

```python
## una lista de lenguas de Sudamérica

glottocode_lenguas_SA = list(glottolog['glottocode'])
```

```python
glottocode_lenguas_SA[:10]
```

**Pregunta 2.** ¿Cómo organizamos la lista de lenguas de Sudamérica en familias lingüísticas?

```python
## archivo descargado de glottolog
languoid = pd.read_csv('languoid.csv')
```

```python
languoid
```

```python
## filtremos la columna "id" con las lenguas de la lista glottocode_lenguas_SA

languoid = languoid[languoid['id'].isin(glottocode_lenguas_SA)]
```

```python
languoid
```

```python
## arreglemos los NaN

## reemplazamos con un string 
languoid.fillna('unk', inplace=True)
```

```python
languoid
```

```python
## construyamos un diccionario glottocode:glottocode_familia

dict_glottocode_fam = {}

## usamos las dos primeras columnas

## recorremos las lenguas de la primera columna
for i in range(len(languoid['id'])):
    ## accedemos a la lengua
    lengua = list(languoid['id'])[i]
    ## accedemos a la familia
    familia = list(languoid['family_id'])[i]
    ## guardamos la información en el diccionario
    dict_glottocode_fam[lengua] = familia
```

```python
dict_glottocode_fam
```

```python
## identifiquemos las familias únicas

## lista vacía
familias_unicas = []

## recorremos las familias
for familia in dict_glottocode_fam.values():
    ## si no está, la guardamos
    if familia not in familias_unicas:
        ## guardamos
        familias_unicas += [familia]
```

```python
## juntemos las lenguas con la misma familia

## diccionario vacío
dict_fam_lenguas = {}

## definimos pares familia:[]
## recorremos las familias
for familia in familias_unicas:
    ## definimos los pares
    dict_fam_lenguas[familia] = []
    
## recorremos las lenguas
for lengua in dict_glottocode_fam.keys():
    ## guardamos las lenguas en las listas respectivas
    ## identificamos la familia
    familia = dict_glottocode_fam[lengua]
    ## guardamos
    dict_fam_lenguas[familia] += [lengua]
```

```python
dict_fam_lenguas
```

### Sails!

```python
## ¿Y si queremos descargar los datos de Sails y leerlos con Python? Usamos la famosa librería pandas :)

## importamos la librería
import pandas as pd

## leemos los tres archivos de la carpeta "datos sails"
## usamos el separador, tab
## usamos el nombre de los archivos

## argex5-1
df_argex5_1 = pd.read_csv('datos sails/argex5-1.txt', sep='\t')

## argex5-2
df_argex5_2 = pd.read_csv('datos sails/argex5-2.txt', sep='\t')

## argex5-3
df_argex5_3 = pd.read_csv('datos sails/argex5-3.txt', sep='\t')
```

```python
df_argex5_2
```

```python
## ¿Cómo podríamos usar diccionarios para representar esta información?

## un detalle sobre pandas dataframes
## con list() transformamos una columna en lista
lenguas = list(df_argex5_1['name'])
```

```python
lenguas[:10]
```

```python
## valores para los distintos rasgos

## argex5-1
valores_argex5_1 = list(df_argex5_1['value'])

## argex5-2
valores_argex5_2 = list(df_argex5_2['value'])

## argex5-3
valores_argex5_3 = list(df_argex5_3['value'])
```

```python
valores_argex5_1[:10]
```

```python
valores_argex5_2[:10]
```

```python
valores_argex5_3[:10]
```

```python
## ¿Qué diccionarios podrían construirse? 

## idea 1: tres diccionarios - argex5_X_dict = {lengua1:valor, lengua2:valor, ...}

## diccionario vacío rasgo argex5-1
argex5_1_dict = {}

## recorremos las lenguas
for i in range(len(lenguas)):
    ## extraemos el valor
    valor = valores_argex5_1[i]
    ## guardamos en el diccionario
    argex5_1_dict[lenguas[i]] = valor
```

```python
argex5_1_dict
```

## mini desafío 2
En este mini desafío, usted y su equipo deben construir un diccionario cuyas keys sean la lista **familias_unicas**, y cuyos values sean listas con todas las lenguas de **lenguas** que pertenezcan a esa familia. **Recomendación.** Usen el diccionario **lengua_familia**. Noten que en la definición del nuevo diccionario usamos la lista construida en el mini desafío 1. 

```python
familia_lista_lenguas = {}
```

```python
## recorremos la lista de familias únicas
for familia in familias_unicas:
    
    ## definimos una lista vacía: aqui guardamos las lenguas asociadas a esta familia
    lista_lenguas = []
    ## recorremos todas las lenguas
    for lengua in lengua_familia.keys():
        ## condición
        if familia == lengua_familia[lengua]:
            ## guardamos las lenguas
            lista_lenguas += [lengua]
    familia_lista_lenguas[familia] = lista_lenguas
```

```python
familia_lista_lenguas
```

```python
## inicializamos el diccionario usando esta forma

## primero parte vacío
familia_lista_lenguas = {}

## recorremos la lista familias_unicas 
for familia in familias_unicas:
    ## asignamos listas vacías
    familia_lista_lenguas[familia] = []
```

```python
## recorrer las lenguas
for lengua in lengua_familia.keys():
    ## accedemos a la familia
    familia = lengua_familia[lengua]
    ## guardamos
    familia_lista_lenguas[familia] += [lengua]
```

```python
familia_lista_lenguas
```

## Preguntas a los datos de [Glottolog](https://glottolog.org/)

```python
## ¿Cuántas lenguas hay por familia? Respondamos usando un diccionario del tipo familia:#lenguas

## diccionario con el resultado
familia_numero_lenguas = {}

## recorremos el diccionario familia_lista_lenguas
for familia in familia_lista_lenguas.keys():
    ## ¿Cuántas lenguas?
    n = len(familia_lista_lenguas[familia])
    ## almacenamos esta información
    familia_numero_lenguas[familia] = n
```

```python
familia_numero_lenguas
```

```python
## ¿Cuántas lenguas tienen en promedio las familias analizadas? Usemos una librería famosa

## la librería famosa, importada como np
import numpy as np

## necesitamos promediar los values de la lista familia_lista_lenguas
promedio = np.mean(list(familia_numero_lenguas.values()))

## necesitamos promediar los values de la lista familia_lista_lenguas
std = np.std(list(familia_numero_lenguas.values()))
```

```python
np.round(promedio,1),np.round(std,1)
```

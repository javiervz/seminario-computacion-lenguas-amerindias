### ¿Cómo construimos un mapa con las lenguas de Sudamérica que aparecen en Wals?

```python
## lenguas de wals https://github.com/cldf-datasets/wals/tree/master/cldf
languages_wals = pd.read_csv('languages_wals.csv')
```

```python
languages_wals
```

```python
## filtramos por lenguas de Sudamérica
languages_wals = languages_wals[languages_wals['Macroarea'] == 'South America']
```

```python
languages_wals
```

```python
## extraemos tres listas: lenguas, latitud, longitud

## lenguas
lenguas = list(languages_wals['Name'])

## latitud
latitud = list(languages_wals['Latitude'])

## longitud
longitud = list(languages_wals['Longitude'])
```

```python
lenguas[:5], latitud[:5], longitud[:5]
```

```python
## construyamos un diccionario lengua:[latitud, longitud]
diccionario_loc = {}

## recorremos los índices
for i in range(len(lenguas)):
    ## lengua
    lengua = lenguas[i]
    
    ## latitud
    lat = latitud[i]
    
    ## longitud
    lon = longitud[i]
    
    ## guardamos
    diccionario_loc[lengua] = [lat, lon]
```

```python
diccionario_loc
```

### ¿Cómo filtramos por las lenguas que aparecen dentro de las fronteras de Perú?

```python
!pip install Shapely
```

```python
## usamos nuestro diccionario

diccionario_loc['Arabela']
```

```python
## filtramos el mundo para que quede solo Perú

world = world[world['name']=='Peru']
```

```python
world
```

```python
## probemos para una lengua

from shapely.geometry import Point, Polygon

# Create Point objects
p1 = Point(diccionario_loc['Arabela'][1], diccionario_loc['Arabela'][0])

Point(diccionario_loc['Arabela'][1], diccionario_loc['Arabela'][0]).within(list(world['geometry'])[0])
```

```python
## diccionario con las lenguas de Perú
diccionario_peru = {}

## filtramos
for lengua in diccionario_loc.keys():
    ## condición: la lengua está dentro de las fronteras
    if Point(diccionario_loc[lengua][1], diccionario_loc[lengua][0]).within(list(world['geometry'])[0]):
        ## guardamos
        diccionario_peru[lengua] = diccionario_loc[lengua]
```

```python
len(diccionario_peru)
```

```python
diccionario_peru
```

```python
## nuevas listas
latitud = []
longitud = []

for lengua in diccionario_peru.keys():
    latitud += [diccionario_peru[lengua][0]]
    longitud += [diccionario_peru[lengua][1]]
```

```python
## otra versión: etiquetas

## conviene definir fig
fig, ax = plt.subplots(dpi=800)

## título
plt.title('Lenguas de Perú anotadas en Wals',fontsize=5)

## colores
world.plot(ax=ax, color='honeydew', edgecolor='k',linewidth=0.2)

## usamos annotate
for language in diccionario_peru.keys():
    ax.annotate(language, color='k', xy=(diccionario_loc[language][1], diccionario_loc[language][0]), fontsize=2., alpha=0.85)

ax.plot(longitud,latitud, marker='o',color='orange',markersize=4.5,markeredgewidth=0.1,markeredgecolor='k',alpha=0.75,linewidth=0,fillstyle='full',clip_on=True)

#plt.legend(loc='best',fontsize=3)
ax.set_yticks([])
ax.set_xticks([])
plt.savefig('map_peru.jpg', format='jpg', transparent=True, bbox_inches='tight',dpi=800)
plt.show()
```

### ¿Cómo podemos visualizar la cobertura en el mapa?

```python
## buscamos los códigos iso de las lenguas de diccionario_peru

languages_peru = languages_wals[languages_wals['Name'].isin(list(diccionario_peru.keys()))]
```

```python
len(languages_peru)
```

```python
## construimos un diccionario id:name

## id + names
id_codes = list(languages_peru['ID'])
names = list(languages_peru['Name'])

## id_names
id_names = {}

## recorremos índices
for i in range(len(id_codes)):
    ## id
    id_ = id_codes[i]
    ## nombre
    name = names[i]
    ## guardamos
    id_names[id_] = name
```

```python
len(id_names)
```

```python
## rasgos
rasgos_wals = pd.read_csv('values_wals.csv')
```

```python
rasgos_wals
```

```python
## filtramos los rasgos por iso

rasgos_wals = rasgos_wals[rasgos_wals['Language_ID'].isin(id_names.keys())]
```

```python
rasgos_wals
```

```python
## lista de repeticiones

rasgos = list(rasgos_wals['Language_ID'])
```

```python
len(set(rasgos))
```

```python
## ¿Cuántas veces aparece cada lengua?
## usemos una librería

from collections import Counter

## usamos la magia
diccionario_cuentas = dict(Counter(rasgos))
```

```python
diccionario_cuentas 
```

```python
## transformamos a nombres
rasgos_peru = {}

## recorremos keys
for id_ in diccionario_cuentas.keys():
    ## usamos el diccionario id:name
    rasgos_peru[id_names[id_]] = diccionario_cuentas[id_]
```

```python
rasgos_peru
```

```python
## ahora volvemos al mapa

## conviene definir fig
fig, ax = plt.subplots(dpi=800)

## título
plt.title('Número de rasgos para lenguas del Perú en Wals',fontsize=5)

## colores
world.plot(ax=ax, color='honeydew', edgecolor='k',linewidth=0.2)

## usamos annotate
for language in diccionario_peru.keys():
    ax.annotate(language, color='k', xy=(diccionario_loc[language][1], diccionario_loc[language][0]), fontsize=2., alpha=0.85)

ax.scatter(longitud,latitud, marker='o',color='orange',s=list(rasgos_peru.values()),alpha=0.75,linewidth=0,clip_on=True)

#plt.legend(loc='best',fontsize=3)
ax.set_yticks([])
ax.set_xticks([])
plt.savefig('map_peru_rasgos.jpg', format='jpg', transparent=True, bbox_inches='tight',dpi=800)
plt.show()
```

```python
## ahora volvemos al mapa: colormap

import matplotlib.cm as cm

## conviene definir fig
fig, ax = plt.subplots(dpi=800)

## título
plt.title('Número de rasgos para lenguas del Perú en Wals',fontsize=5)

## colores
world.plot(ax=ax, color='honeydew', edgecolor='k',linewidth=0.2)

## usamos annotate
for language in diccionario_peru.keys():
    ax.annotate(language, color='k', xy=(diccionario_loc[language][1], diccionario_loc[language][0]), fontsize=2., alpha=0.85)

ax.scatter(longitud,latitud, marker='o',c=list(rasgos_peru.values()),s=12,alpha=0.75,linewidth=0,clip_on=True, cmap='autumn')

#plt.legend(loc='best',fontsize=3)
ax.set_yticks([])
ax.set_xticks([])
plt.savefig('map_peru_rasgos.jpg', format='jpg', transparent=True, bbox_inches='tight',dpi=800)
plt.show()
```

```python
## mostremos algunas familias

## extraemos la lista de familias
familias = list(languages_wals['Family'])
```

```python
## construyamos un diccionario lengua:familia
lengua_familia = {}

## recorremos los índices
for i in range(len(lenguas)):
    ## lengua
    lengua = lenguas[i]
    
    ## familia
    familia = familias[i]
    
    ## guardamos
    lengua_familia[lengua] = familia
```

```python
lengua_familia
```

```python
## otra versión: familias

## conviene definir fig
fig, ax = plt.subplots(dpi=800)

## título
plt.title('Lenguas anotadas en Wals',fontsize=5)

## colores
world.plot(ax=ax, color='honeydew', edgecolor='k',linewidth=0.2)

## usamos annotate
for language in diccionario_loc.keys():
    ## familia
    familia = lengua_familia[language]
    ## anotamos algunas familias
    if familia in ['Pano-Tacanan']:
        ax.annotate(familia, color='k', xy=(diccionario_loc[language][1], diccionario_loc[language][0]), fontsize=2., alpha=0.85)

ax.plot(longitud,latitud, marker='o',color='orange',markersize=2.5,markeredgewidth=0.1,markeredgecolor='k',alpha=0.5,linewidth=0,fillstyle='full',clip_on=True)

#plt.legend(loc='best',fontsize=3)
ax.set_yticks([])
ax.set_xticks([])
plt.savefig('map_SA_familias.jpg', format='jpg', transparent=True, bbox_inches='tight',dpi=800)
plt.show()
```

```python
## otra versión: familias 2

## conviene definir fig
fig, ax = plt.subplots(dpi=800)

## título
plt.title('Lenguas anotadas en Wals',fontsize=5)

## colores
world.plot(ax=ax, color='honeydew', edgecolor='k',linewidth=0.2)

## usamos colores para las familias
## lon, lat para Pano-Tacanan
lon_PT = []
lat_PT = []

## índices lenguas Pano-Tacanan
indices_PT = []

## recorremos las lenguas
for i in range(len(lenguas)):
    ## condición Pano-Tacanan
    if lengua_familia[lenguas[i]] == 'Pano-Tacanan':
        ## guardamos
        indices_PT += [i]
        
## dividimos los datos de latitud/longitud en dos
latitud_PT = []
longitud_PT = []

latitud_non_PT = []
longitud_non_PT = []

## recorremos los índices
for i in range(len(lenguas)):
    ## casos
    if i in indices_PT:
        ## loc PT
        latitud_PT += [latitud[i]]
        longitud_PT += [longitud[i]]
    else:
        ## loc non PT
        latitud_non_PT += [latitud[i]]
        longitud_non_PT += [longitud[i]]
        
ax.plot(longitud_PT, latitud_PT, marker='o',color='orange',markersize=2.5,markeredgewidth=0.1,markeredgecolor='k',alpha=0.75,linewidth=0,fillstyle='full',clip_on=True, label='Pano-Tacanan')
ax.plot(longitud_non_PT, latitud_non_PT, marker='o',color='blue',markersize=2.5,markeredgewidth=0.1,markeredgecolor='k',alpha=0.75,linewidth=0,fillstyle='full',clip_on=True, label='otras familias')

plt.legend(loc='best',fontsize=3)
ax.set_yticks([])
ax.set_xticks([])
plt.savefig('map_SA_familias_colors.jpg', format='jpg', transparent=True, bbox_inches='tight',dpi=800)
plt.show()
```
